```python
X = df.drop(columns=['Name', 'Hired'])
y = df['Hired']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

Logistic Regression

```python
scaler = StandardScaler()

# Fit only on training data, then transform both train and test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
hired_logreg = LogisticRegression(class_weight='balanced', random_state=42, max_iter=10000)
hired_logreg.fit(X_train_scaled, y_train)
accuracy = accuracy_score(y_test, hired_logreg.predict(X_test_scaled))
print(f"Logistic Regression model accuracy: {accuracy:.2f}%")
y_pred_logreg = hired_logreg.predict(X_test_scaled)
print(classification_report(y_test, y_pred_logreg))
```

```python
logRegConfusionMatrix = confusion_matrix(y_test, y_pred_logreg)
logRegCM = ConfusionMatrixDisplay(confusion_matrix=logRegConfusionMatrix, display_labels=['Not Hired', 'Hired'])
fig, ax = plt.subplots(figsize=(6,6))
logRegCM.plot(ax=ax, cmap='Blues', colorbar=False)
ax.set_title("Confusion Matrix for Logistic Regression")
plt.show()
```

RandomForest

```python
hired_rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
hired_rf_classifier.fit(X_train_scaled, y_train)
y_pred_rf = hired_rf_classifier.predict(X_test_scaled)
accuracy = accuracy_score(y_test, hired_rf_classifier.predict(X_test_scaled))
print(f"RandomForest model accuracy: {accuracy:.2f}%")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    hired_rf_classifier,  
    X_train_scaled,        
    y_train,               
    cv=5,                  
    scoring='accuracy'     
)

print(f"Cross-validation scores: {cv_scores}")
print(f"Average CV Accuracy: {cv_scores.mean():.2f}")
```

```python
rfConfusionMatrix = confusion_matrix(y_test, y_pred_rf)
rfCM = ConfusionMatrixDisplay(confusion_matrix=rfConfusionMatrix, display_labels=['Not Hired', 'Hired'])
fig, ax = plt.subplots(figsize=(6,6))
logRegCM.plot(ax=ax, cmap='Blues', colorbar=False)
ax.set_title("Confusion Matrix for RandomForest")
plt.show()
```

Feature Importance

```python
feature_names = X.columns
```

```python
feature_names
```

```python
importances = hired_rf_classifier.feature_importances_
feature_imp_df = pd.DataFrame({'Feature': feature_names, 'Gini Importance': importances}).sort_values('Gini Importance',ascending=False)
print(feature_imp_df)
```

```python
plt.figure(figsize=(8, 4))
plt.barh(feature_names, importances, color='skyblue')
plt.xlabel('Gini Importance')
plt.title('Feature Importance - Gini Importance')
plt.gca().invert_yaxis()
plt.show()
```

```python
importances = []
initial_accuracy = accuracy_score(y_test, hired_rf_classifier.predict(X_test_scaled)) # initial accuracy
for i in range(X.shape[1]):
    X_test_copy = X_test_scaled.copy()
    np.random.shuffle(X_test_copy[:, i])
    shuff_accuracy = accuracy_score(y_test, hired_rf_classifier.predict(X_test_copy))
    importances.append(initial_accuracy - shuff_accuracy)
    
accuracy_df = pd.DataFrame({'Feature': feature_names, 'Decrease in Accuracy': importances}).sort_values('Decrease in Accuracy', ascending=False)
print(accuracy_df)
```

```python
plt.figure(figsize=(8, 4))
plt.barh(feature_names, importances, color='skyblue')
plt.xlabel('Mean Decrease Accuracy')
plt.title('Feature Importance - Mean Decrease Accuracy')
plt.gca().invert_yaxis() 
plt.show()
```

```python
from sklearn.inspection import permutation_importance
result = permutation_importance(hired_rf_classifier, X_test_scaled, y_test, n_repeats=30, random_state=42)
perm_imp_df = pd.DataFrame({'Feature': feature_names, 'Permutation Importance': result.importances_mean}).sort_values('Permutation Importance', ascending=False)
print(perm_imp_df)
```

```python
plt.figure(figsize=(10, 4))
plt.bar(feature_names, result.importances_mean)
plt.xticks(rotation=45, ha='right') 
plt.xlabel("Features")
plt.ylabel("Permutation Importance")
plt.title("Feature Importance - Permutation Importance")
plt.tight_layout()
plt.show()
```

Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_classification
```

GridSearchCV

```python
X = df[['Experience (Years)', 'Projects Count', 'Skill Count', 'Salary Expectation ($)', 'Has Certification']]
y = df['Hired']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

hired_logreg = LogisticRegression(class_weight='balanced', random_state=42, max_iter=10000)
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}

logreg_cv = GridSearchCV(hired_logreg, param_grid, cv=5)
logreg_cv.fit(X_train_scaled, y_train)
y_pred = logreg_cv.predict(X_test_scaled)

print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_))
print("Best score is {}".format(logreg_cv.best_score_))
print()
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
logReggGSConfusionMatrix = confusion_matrix(y_test, y_pred)
logRegCM = ConfusionMatrixDisplay(confusion_matrix=logReggGSConfusionMatrix, display_labels=['Not Hired', 'Hired'])
fig, ax = plt.subplots(figsize=(6,6))
logRegCM.plot(ax=ax, cmap='Blues', colorbar=False)
ax.set_title("Confusion Matrix for Logistic Regression")
plt.show()
```
